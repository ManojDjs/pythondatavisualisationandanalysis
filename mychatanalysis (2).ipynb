```python
from pymongo import MongoClient
import re
import matplotlib.pyplot as plt
client = MongoClient('mongodb://localhost:27017/')
if(client):
   print("connected")
db = client.JOBS
collection2=db.chatbot
doc=collection2.find({})
dateslist=[]
for i in doc:
        #print(i['text'])
        msg=i['text'].split(',')
        datemsg=msg[0]
        match = re.search(r'(\d+/\d+/\d+)', datemsg)
        if match:
            dateslist.append(datemsg)
dic={}
for i in dateslist:
       dic[i]=dateslist.count(i)
dates=list(dic.keys())   
count=list(dic.values())
print(dates)
print(count)
from datetime import datetime
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as pyplot
data={'date':dates,'count':count}
df=pd.DataFrame(data,columns=['date','count'])
print(df)
```

```python
from datetime import datetime
df['date']=[str(i).replace('0','') for i in df['date']]
print(type(df['date'][0]))
df['date']=[datetime.strptime(i, '%d/%m/%y') for i in df['date']]
#df['date']=[datetime.strptime(d, '%m/%d/%y') for d in df['date']]
print(df['date'])
```

```python
plt.figure(figsize=(50,20))
plt.bar(df['date'],df['count'])
plt.rc('xtick',labelsize=25)
plt.rc('ytick',labelsize=38)
plt.gcf().autofmt_xdate()
plt.show()
```
