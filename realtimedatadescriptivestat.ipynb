```python
import pandas as pd
import numpy as np
from pandas import DataFrame,Series
df=pd.read_csv(r'C:\Users\behar\Desktop\pdfdocs\stat.csv')
print(df)
#here i plot i want to plot graph between area and price 
#which is abnormal because  price will be always depends on the place we choose 
```

```python
print(df.fillna(''))
```

```python
print(df['price'].replace('$',' '))
```

```python
arealist=df['area'].tolist()
pricelist=df['price'].tolist()
pricelist=[x.replace('$','') for x in pricelist]
pricelist=[x.replace(',','') for x in pricelist]
print(pricelist)
newareapricedataframe=pd.DataFrame({'area':arealist,'price':pricelist})
print(newareapricedataframe)
```

```python
sorteddf=newareapricedataframe.sort_values(by='price')
print(sorteddf)
```

```python
arealist1=sorteddf['area'].tolist()
pricelist1=sorteddf['price'].tolist()
import matplotlib.pyplot as plt
plt.figure(figsize=(20,50))
print(plt.title('area in foot to its price in dollars'))
plt.ylabel('price in dollars')
plt.xlabel('area in sqfeet')
plt.plot(arealist1, pricelist1)
plt.axis([min(arealist1),max(arealist1),min(pricelist1), max(pricelist1)])
print(plt.show())
```
