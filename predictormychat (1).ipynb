```python
# Timedelta function demonstration  
from datetime import datetime, timedelta 
# Using current time 
ini_time_for_now = datetime.now() 
# printing initial_date 
print ("initial_date", ini_time_for_now)  
# Some another datetime 
#you can add hours day and any time realted data to time delta
new_final_time = ini_time_for_now + timedelta(hours = 2)
# printing new final_date 
print ("new_final_time",new_final_time.time()) 
  
```

```python
from pymongo import MongoClient
import re
import matplotlib.pyplot as plt
client = MongoClient('mongodb://localhost:27017/')
if(client):
   print("connected")
db = client.JOBS
collection2=db.chatbot
doc=collection2.find({})
dateslist=[]
for i in doc:
    #print(i['text'])
    msg=i['text'].split('-')
    datemsg=msg[0]
    match = re.search(r'(\d+/\d+/\d+)', datemsg)
    if match:
        dateslist.append(datemsg)
dic={}
for i in dateslist:
	dic[i]=dateslist.count(i)
dates=list(dic.keys())
count=list(dic.values())
print(dates)
print(count)
sum=0
for i in count:
    sum=sum+1
print(sum)    
```

```python
data={'8:00 am':0,'9:00 am':0,'10:00 am':0,'11:00 am':0,'12:00 pm':0,'1:00 pm':0,'2:00 pm':0,'3:00 pm':0,'4:00 pm':0,'5:00 pm':0,
     '6:00 pm':0,'7:00 pm':0,'8:00 pm':0,'9:00 pm':0,'10:00 pm':0,'11:00 pm':0}
print(data.keys())
```

```python
import datetime 
i='1:00 pm'
print(datetime.datetime.strptime(i,'%I:%M %p'))
'''def convert(i):
       d=datetime.datetime.strptime(i,'%I:%M %p')
       print(d.hour)
    #convert(i)'''
dates=[datetime.datetime.strptime(i,'%d/%m/%y, %I:%M %p ') for i in dates]
for j in data.keys():
    print(j)
    counts=0
    compare=datetime.datetime.strptime(j,'%I:%M %p')
    h=compare.hour
    #print(compare.hour)
    for i in dates:
        if i.hour==h:
            counts=counts+1
    data[j]=counts        
    print(counts)    
print(data)   
```

```python
import matplotlib.pyplot as plt
title_string=('texts per hour')
plt.figure(figsize=(50,20))
plt.suptitle(title_string, y=1.0, fontsize=45)
plt.bar(data.keys(), data.values())
plt.rc('xtick',labelsize=25)
plt.rc('ytick',labelsize=38)
plt.gcf().autofmt_xdate()
plt.xlabel('time lapse',fontsize=50)
plt.ylabel('messages per each hour',fontsize=50)
plt.show()
```
